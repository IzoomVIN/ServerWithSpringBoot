```python
import requests
import json
import pandas as pd
```

```python
class TestFullREST(object):
    __url = "http://localhost:666/SpringBootOTS/api/"
    
    def __init__(self):
        pass
    
    '''return dataFrame table'''
    def getAllData(self):
        url = self.__url + "get/all"
        
        response = requests.get(url)
        data = response.json()

        headers = list(data[0].keys()) 
        myList = self._fromJSONToList(data) 
        return pd.DataFrame(data=myList, columns=headers)
    
    '''return dataFrame table or string with response'''
    def getDataForTheYear(self, year):
        url = self.__url + "get/%s"%year
        
        response = requests.get(url)
        data = response.json()
        
        if type(data) == str:
            return data
        
        headers = list(data[0].keys()) 
        myList = self._fromJSONToList(data) 
        return pd.DataFrame(data=myList, columns=headers)
    
    '''return dataFrame table with one row'''
    def getRow(self, country, year, sex, age):
        url = self.__url + "row/get"
        params = {"country" : country, "year" : year, "sex" : sex, "age" : age}
        
        response = requests.get(url, params=params)
        data = response.json()

        headers = list(data.keys()) 
        myList = self._fromJSONToList(data)
        return pd.DataFrame(data= [myList], columns= headers)
    
    '''return String of result'''
    def addRow(self, row):
        url = self.__url + "row/add"
        response = requests.post(url, data=json.dumps(row)) 
        return response.text;
    
    '''return String of result'''
    def updateRow(self, row):
        url = self.__url + "row/update"
        response = requests.post(url, data=json.dumps(row)) 
        return response.text;
    
    '''return String of result'''
    def deleteRow(self, country, year, sex, age):
        url = self.__url + "row/delete"
        params = {"country" : country, "year" : year, "sex" : sex, "age" : age}
        
        response = requests.delete(url, params=params)
        return response.text
    
    def _fromJSONToList (self, data):
        result = []
        if type(data) == list:
            for d in data:
                line = []
                for atr in d.values():
                    line.append(atr)
                result.append(line)
        elif type(data) == dict:
            for atr in data.values():
                result.append(atr)
            
        return result
```
